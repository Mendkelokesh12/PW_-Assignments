Q1. What is a database? Differentiate between SQL and NoSQL databases.

```python
print(''' A database is a structured collection of data that is stored and accessed electronically. 
Databases are crucial for storing, retrieving, and managing large amounts of information efficiently. 
They are used in various applications, ranging from websites and mobile apps to enterprise-level systems.


''')

print(''' SQL Databases:

    Structure:
    SQL databases are relational databases, meaning they use a structured query language for defining and manipulating data.
    They are based on a schema that defines the structure of data, including tables, fields, and the relationships between them.
    
    ACID Properties: 
    SQL databases strongly adhere to ACID (Atomicity, Consistency, Isolation, Durability) properties, ensuring reliable transactions and data integrity.
    
    Examples: 
    Popular SQL databases include MySQL, PostgreSQL, Oracle, and Microsoft SQL Server.
    
    Use Cases: 
    Ideal for applications that require complex queries, transactions, and where data integrity is critical.
    
    ''')

print('''  NoSQL Databases:

    Structure: 
    NoSQL databases are non-relational or distributed databases.They are schema-less, allowing for the storage of unstructured and semi-structured data.
    They come in various forms, including key-value stores, document stores, column stores, and graph databases.
    
    Scalability:
    They are known for their high scalability and flexibility in handling large volumes of data and high traffic loads.
    
    Consistency Models: 
    NoSQL databases often use eventual consistency models, which can be more suitable for distributed systems where immediate consistency
    is not as critical.
    
    Examples: 
    Examples of NoSQL databases include MongoDB, Cassandra, Redis, and Neo4j.
    
    Use Cases:
    They are well-suited for big data applications, real-time analytics, and situations where the data structure is 
    not clearly defined or can change over time.''')
```

Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

```python
print(''' DDL stands for Data Definition Language. It is a subset of SQL (Structured Query Language) used to define and
modify the structure of database objects in a database. These objects include tables, indexes, constraints, 
and database schemas.DDL statements are used to create, modify, or delete these structures, 
but they do not deal with data manipulation within the tables (which is handled by Data Manipulation Language, or DML).
''')

print(''' CREATE:
This command is used to create a new object in the database, such as a table, index, or view. For example, to create a new table:

CREATE TABLE Students (
    StudentID int,
    Name varchar(255),
    Age int,
    GradeLevel varchar(50)
);

''')

print(''' DROP:
This command is used to delete an existing database object, like a table or a database itself.
For example:  

DROP TABLE Students;

''')

print(''' ALTER: 
This command is used to modify an existing database object. For example, to add a new column to a table:

ALTER TABLE Students ADD Email varchar(255);

''')

print(''' TRUNCATE: 
This command is used to remove all records from a table, but it does not delete the table itself.
It is like a fast, efficient DELETE command without a WHERE clause.
For example: 

TRUNCATE TABLE Students;

''')
```

Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

```python
print(''' DML stands for Data Manipulation Language. It is a subset of SQL (Structured Query Language) used for adding (inserting), 
modifying (updating), and deleting data in a database. DML commands do not modify the schema or structure of the database itself, 
but they are used to manipulate the data within the tables.
Key DML commands include INSERT, UPDATE, and DELETE
''')

print(''' 
INSERT: 
This command is used to add new rows (records) to a table.
For example: 

INSERT INTO Students (StudentID, Name, Age, GradeLevel)
VALUES (1, 'John Doe', 16, '10th');
''')

print(''' 
UPDATE: 
This command is used to modify existing data in a table. It is typically used in conjunction with a 
WHERE clause to specify which rows should be updated.
For example:

UPDATE Students SET Age = 17 WHERE StudentID = 1;

''')

print('''  
DELETE: 
This command is used to remove rows from a table. Like UPDATE, it is often used with a
WHERE clause to specify which rows to delete. 
For example:

DELETE FROM Students WHERE StudentID = 1;

''')
```

Q5. Explain Primary Key and Foreign Key.

```python
print(''' Primary Key:

    Definition:
    A Primary Key is a column or a set of columns in a table that uniquely identifies each row in that table. 
    No two rows can have the same primary key value, and a primary key cannot be null.
    
    Purpose: 
    The primary purpose of a primary key is to provide a unique identifier for each record in a table.
    This ensures that each row can be uniquely identified and referenced.
    
    Example:
    In a Students table, StudentID could be a primary key. Each student would have a unique StudentID,
    differentiating them from other students in the table. 
    
    
    CREATE TABLE Students (StudentID int PRIMARY KEY, Name varchar(255), Age int);
''')

print('''
Foreign Key:

    Definition:
    A Foreign Key is a column or a set of columns in one table that references the primary key columns of another table.
    The purpose of a foreign key is to maintain referential integrity between two related tables.
    
    Purpose:
    A foreign key establishes a link between two tables, ensuring that the data referenced in one table exists in the other table. 
    This is crucial for maintaining consistent and accurate data across multiple tables.
    
    Example:
    If there is a Courses table and a Students table, and each course is taken by multiple students, 
    a StudentID column in the Courses table could act as a foreign key referencing the StudentID primary key in the Students table
    
    CREATE TABLE Courses (CourseID int PRIMARY KEY, CourseName varchar(255), StudentID int, FOREIGN KEY (StudentID) REFERENCES Students(StudentID));
''')
```

Q6. Write a python code to connect MySQL to python. Explain the cursor() and execute() method.

```python
print(''' import mysql.connector

# Establish a database connection
connection = mysql.connector.connect(
    host='hostname',        # your host, usually localhost
    user='username',        # your username
    password='password',    # your password
    database='databasename' # name of the database
)

# Create a cursor object
cursor = connection.cursor()

# Execute a query
cursor.execute("SELECT * FROM your_table_name")

# Fetch the results
results = cursor.fetchall()

for row in results:
    print(row)

# Close the cursor and connection
cursor.close()
connection.close()


cursor() Method:
After establishing a connection to the database, you create a cursor object using the cursor() method of the connection object. 
The cursor is used to execute SQL commands. Think of the cursor as a pointer that enables traversal over the records in the database.

execute() Method:
The execute() method of the cursor object is used to execute SQL commands. In this example,
cursor.execute("SELECT * FROM your_table_name") executes a SQL query to select all records from the specified table.''')
```

Q7. Give the order of execution of SQL clauses in an SQL query.

```python
print(''' 
The order of execution of SQL clauses in an SQL query is important for understanding how SQL queries are processed by the database server.
This order determines how the result set is built and returned. The actual order in which SQL clauses are written in a query is different 
from the order in which the database engine processes them. Here's the typical order of execution:

    FROM clause:
    The execution starts with the FROM clause, where the database determines which tables to scan. If there are joins, they are performed here.

    JOIN clauses:
    If the query involves any joins, they are processed at this stage. The database combines rows from different tables based on the join conditions.

    WHERE clause: 
    The WHERE clause is applied next. This filters the rows returned by the FROM and JOIN clauses based on the specified condition. 
    Only the rows that meet the condition are included in the result set.

    GROUP BY clause:
    If the query includes a GROUP BY clause, the rows that survived the WHERE filter are then grouped based on the specified columns. 
    This is used in conjunction with aggregate functions (like COUNT, SUM, AVG, etc.).

    HAVING clause: 
    The HAVING clause is similar to the WHERE clause, but it is applied to groups rather than individual rows. 
    It filters groups created by the GROUP BY clause.

    SELECT clause:
    At this point, the database engine processes the SELECT clause, selecting the specified columns from the rows or groups. 
    Any expressions in the SELECT clause are calculated at this stage.

    DISTINCT keyword: 
    If the SELECT clause includes the DISTINCT keyword, duplicate rows are removed from the set.

    ORDER BY clause: 
    The ORDER BY clause sorts the resulting rows or groups in ascending or descending order as specified. 
    This is usually the last step before the results are returned to the user.

    LIMIT / OFFSET clause: 
    If present, the LIMIT and OFFSET clauses are applied last. LIMIT restricts the number of rows returned,
    and OFFSET specifies the number of rows to skip before starting to return rows 
    ''')
```
