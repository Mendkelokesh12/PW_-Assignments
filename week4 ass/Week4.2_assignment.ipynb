What is Abstraction in OOps? Explain with an example.

```python
print('''Abstraction is one of the fundamental principles of Object-Oriented Programming (OOP) 
      that allows you to hide the complex implementation details and show only the essential features of an object.
      It helps in managing the complexity of a system by providing a simplified and clear interface''')

from abc import ABC, abstractmethod

class shape(ABC):
    
    @abstractmethod
    def calcu_area(self):
        pass
    
    @abstractmethod
    def dispaly(self):
        pass
    
class Circle(shape):
    
    def __init__(self,radius):
        self.r = radius
        
    def calcu_area(self):
        return 3.14 * self.r *self.r
    
    def dispaly(self):
        return "circle With Radius",self.r
    
class Rectangle(shape):
    
    def __init__(self,length,breadth):
        self.l = length
        self.b = breadth
        
    def calcu_area(self):
        return self.l * self.b
    
    def dispaly(self):
        return "Rectangle with length ",self.l, " and width " ,self.b
    
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Area of the circle: {circle.calcu_area()}")
circle.dispaly()
print(f"Area of the rectangle: {rectangle.calcu_area()}")
rectangle.dispaly()
```

Differentiate between Abstraction and Encapsulation. Explain with an example.

```python
print(''' Abstraction:

    Focus: What the object does, not how it does it.
    Mechanism: Abstract classes and methods.
    Purpose: Hide implementation details, simplify interactions, promote reusability.
    
    Example:
    
''')
from abc import ABC, abstractmethod

class Vehicle(ABC):  # Abstract class, hiding implementation details
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def move(self):
        pass

class Car(Vehicle):  # Concrete subclass, providing specific implementations
    def start(self):
        print("Car starting...")

    def move(self):
        print("Car moving on the road.")

class Boat(Vehicle):  # Concrete subclass, providing different implementations
    def start(self):
        print("Boat engine starting...")

    def move(self):
        print("Boat sailing on the water.")




print('''

Encapsulation:

    Focus: Bundling data and related operations within a single unit (object).
    Mechanism: Classes, access modifiers (public, private, protected).
    Purpose: Protect data integrity, control access, promote modularity.
    
    Example:

  ''')
class Account:
    def __init__(self, name, balance):
        self.__name = name   # Private attribute
        self.__balance = balance  # Private attribute

    def get_balance(self):  # Public method to access balance
        return self.__balance

    def deposit(self, amount):  # Public method to modify balance
        self.__balance += amount

    def withdraw(self, amount):  # Public method to modify balance
        if self.__balance >= amount:
            self.__balance -= amount
        else:
            print("Insufficient funds")



```

What is abc module in python? Why is it used?

```python
print('''It stands for "Abstract Base Classes."
It provides tools for defining abstract base classes (ABCs) in Python.
ABCs serve as blueprints for other classes, defining common interfaces and enforcing certain behaviors.

''')
print(''' Defining Abstract Classes:

    Python doesn't have built-in support for abstract classes.
    The abc module enables their creation using the ABCMeta metaclass and @abstractmethod decorator.

Enforcing Interfaces:

    ABCs ensure that subclasses implement specific methods, guaranteeing adherence to a defined interface.
    This promotes code consistency and prevents errors due to missing methods.

Defining Type Hierarchies:

    The abc module supports creating custom type hierarchies, useful for type checking and code organization.
    For example, you can create a hierarchy of numeric types or collection types.

Design Patterns:

    ABCs are often used in design patterns like the Template Method pattern and 
    the Strategy pattern to define common structures and allow for flexible variations.''')
```

How can we achieve data abstraction?

```python
print(''' Abstract Classes and Methods:

    Abstract classes define a blueprint for other classes, specifying essential features and behavior.
    They cannot be instantiated directly.
    Abstract methods are declared within abstract classes but have no implementation.
    Subclasses must implement these methods to provide concrete functionality.

Example: ''')

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        print("Woof!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

        
print('''  Access Modifiers:

    Python doesn't have strict access modifiers like private and public, but it uses conventions:
        Underscore prefix (e.g., _name): Implies private, intended for internal use.
        Double underscore prefix (e.g., __name): Name mangling, making attributes harder to access externally.

Example:''')


class Account:
    def __init__(self, name, balance):
        self._name = name  # Protected attribute
        self.__balance = balance  # More strongly protected attribute

    def get_balance(self):
        return self.__balance

    
print('''  Properties:

    Provide a way to control access to attributes and add logic during attribute access.
    Use the @property decorator to define a method as a property, accessible like an attribute.

Example: ''')


class Person:
    def __init__(self, age):
        self._age = age

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, new_age):
        if new_age >= 0:
            self._age = new_age
        else:
            raise ValueError("Age cannot be negative")


print(''' Special Methods (Dunder Methods):

    Define behaviors for built-in operations like accessing attributes, creating objects, and comparison.
    Offer a level of abstraction for specific functionalities.
    Example:''')



class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
```
