Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

```python
print('''In Object-Oriented Programming (OOP), a class is a blueprint or a template for creating objects,and an object is an instance of a class. 
Classes define the attributes (characteristics) and behaviors (actions or methods) that objects of the class will have. 
Objects, on the other hand, are instances of classes and represent specific instances of the class with their own unique data.\n''')


# Define a class named 'Dog'
class Dog:
    # Class attribute
    species = "Canine"

    # Constructor method
    def __init__(self, name, age):
        # Instance attributes
        self.name = name
        self.age = age

    # Instance method
    def bark(self):
        print(f"{self.name} says Woof!")

    # Instance method
    def describe(self):
        print(f"{self.name} is {self.age} years old.")

# Create objects (instances) of the 'Dog' class
dog1 = Dog(name="Buddy", age=3)
dog2 = Dog(name="Max", age=5)

# Accessing attributes and methods of objects
print(f"{dog1.name} is a {dog1.species}.")
dog1.describe()
dog1.bark()

print(f"{dog2.name} is a {dog2.species}.")
dog2.describe()
dog2.bark()
```

```python
print('''
1 Encapsulation:

          Definition: Encapsulation involves bundling the data (attributes) and the methods (functions) that operate on the data into a single unit,
          known as a class. It restricts access to some of the object's components, protecting its internal state and implementation details. 
          Encapsulation helps in reducing complexity and promoting information hiding.  \n''')

print(''' 
2 Abstraction
          
          Definition: Abstraction is the process of hiding the complex implementation details of an object and exposing only the essential features or functionalities.
          It focuses on simplifying the representation of an object by emphasizing what an object does rather than how it achieves its functionality. 
          Abstraction helps in managing complexity and providing a clear interface for interacting with objects. \n''')

print('''
3 Inheritance:
          Definition: Inheritance is a mechanism that allows a new class (subclass or derived class) to inherit properties and
          behaviors from an existing class (superclass or base class). The subclass can reuse and extend the functionalities of the superclass. 
          Inheritance promotes code reusability, making it easier to create and maintain a hierarchy of related classes.\n''')

print(''' 
4 Polymorphism:

          Definition: Polymorphism allows objects of different classes to be treated as objects of a common base class. 
          It enables a single interface to represent different types of objects, and it can take various forms,
          such as method overloading (multiple methods with the same name but different parameters) and method overriding (redefining a method in a subclass).
          Polymorphism enhances flexibility and extensibility in the design of software systems.''')
```

Explain why the __init__() function is used. Give a suitable example.

```python
print(''' The __init__() function in Python is a special method, also known as a constructor, 
that is automatically called when an object is created from a class. It is used to initialize the attributes or properties of the object. 
The primary purpose of the __init__() method is to set up the initial state of the object by assigning values to its attributes.\n''')

class Car:
    # The __init__ method is the constructor
    def __init__(self, make, model, year):
        # Attributes of the class
        self.make = make
        self.model = model
        self.year = year
        self.is_running = False

    # Method to start the car
    def start_engine(self):
        print(f"The {self.year} {self.make} {self.model}'s engine is now running.")
        self.is_running = True

    # Method to stop the car
    def stop_engine(self):
        print(f"The {self.year} {self.make} {self.model}'s engine is now stopped.")
        self.is_running = False

# Creating instances of the Car class
car1 = Car(make="Toyota", model="Camry", year=2022)
car2 = Car(make="Honda", model="Accord", year=2023)

# Accessing attributes and methods of the objects
print(f"Car 1: {car1.year} {car1.make} {car1.model}")
print(f"Car 2: {car2.year} {car2.make} {car2.model}")

# Starting and stopping the engines
car1.start_engine()
car2.start_engine()

car1.stop_engine()
car2.stop_engine()
```

Why self is used in OOPs?

```python
print(''' 
In Object-Oriented Programming (OOP), self is a conventional name used to represent the instance of a class within its own methods. 
It is the first parameter of any method in a class, including special methods like __init__. 
The purpose of self is to reference the instance of the class and allow the methods to access and 
modify the attributes and behaviors of that specific instance.

Here are some key points about why self is used in OOP:

    Instance Reference:
        self is used as a reference to the instance of the class. When you create an object (instance) of a class, 
        self allows the methods to operate on the specific instance, distinguishing it from other instances of the same class.

    Accessing Attributes:
        Within class methods, you use self to access and modify instance attributes. For example, self.attribute_name refers
        to the attribute of the current instance.

    Invoking Other Methods:
        self is necessary to invoke other methods of the same class. When a method is called within the class, 
        it needs to use self.method_name() to call other methods.

    Creating and Modifying State:
        self is crucial for creating and maintaining the state of the object. It allows methods to modify
        the internal state of the instance, ensuring that changes are specific to that instance.\n
        
        
''')

class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says Woof!")

    def celebrate_birthday(self):
        self.age += 1
        print(f"{self.name} is now {self.age} years old!")

# Creating an instance of the Dog class
my_dog = Dog(name="Buddy", age=3)

# Accessing attributes and invoking methods using 'self'
print(f"My dog's name is {my_dog.name} and it is {my_dog.age} years old.")
my_dog.bark()

# Modifying the state using 'self'
my_dog.celebrate_birthday()
print(f"Now, {my_dog.name} is {my_dog.age} years old.")
```

What is inheritance? Give an example for each type of inheritance.


```python
print(''' Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a new class to inherit the properties and 
behaviors (attributes and methods) of an existing class. The existing class is called the base class or superclass, 
and the new class is called the derived class or subclass. Inheritance promotes code reuse, 
as the subclass can reuse and extend the functionalities of the superclass.\n''')

print('''  Single Inheritance:

    In single inheritance, a subclass inherits from only one superclass.
    
''')


# Single Inheritance Example
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Creating an instance of the derived class
my_dog = Dog()
my_dog.speak()  # Inherited from Animal class
my_dog.bark()   # Specific to Dog class


print(''' \n Multiple Inheritance:

    In multiple inheritance, a subclass inherits from more than one superclass.
    
''')
# Multiple Inheritance Example
class Flyable:
    def fly(self):
        print("Can fly")

class Swimmable:
    def swim(self):
        print("Can swim")

class FlyingFish(Flyable, Swimmable):
    pass

# Creating an instance of the derived class
flying_fish = FlyingFish()
flying_fish.fly()   # Inherited from Flyable class
flying_fish.swim()  # Inherited from Swimmable class

print(''' \n Multilevel Inheritance:

    In multilevel inheritance, a subclass inherits from another subclass.

''')

# Multilevel Inheritance Example
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Poodle(Dog):
    def dance(self):
        print("Poodle dances")

# Creating an instance of the derived class
my_poodle = Poodle()
my_poodle.speak()  # Inherited from Animal class
my_poodle.bark()   # Inherited from Dog class
my_poodle.dance()  # Specific to Poodle class


print('''\n Hierarchical Inheritance:

    In hierarchical inheritance, multiple subclasses inherit from a single superclass.

 ''' )

# Hierarchical Inheritance Example
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Creating instances of the derived classes
my_dog = Dog()
my_cat = Cat()
my_dog.speak()  # Inherited from Animal class
my_dog.bark()   # Specific to Dog class
my_cat.speak()  # Inherited from Animal class
my_cat.meow()   # Specific to Cat class
```
