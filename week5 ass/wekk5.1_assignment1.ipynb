
Q1. What is multithreading in python? Why is it used? Name the module used to handle threads in python.


```python
print(''' Multithreading in Python refers to the concurrent execution of multiple threads within a single program.
A thread is the smallest unit of execution within a process, and multithreading allows multiple threads to run concurrently, 
sharing the same resources such as memory space but having their own execution context. In Python, 
multithreading is used to achieve parallelism and improve the performance of certain types of tasks. 
''')

print(''' Key points about multithreading:
 1. Concurrency,
 2. Parallelism,
 3. Responsiveness,
 4. Resource Sharing,
 5. Use Cases
 ''')
print(''' he threading module is used for handling threads in Python. 
It provides a high-level interface for creating and managing threads. 
The threading module is built on top of the lower-level _thread module, 
and it simplifies the process of working with threads by providing 
abstractions for thread creation, synchronization, and communication. 
''')

import threading
import time

def print_numbers():
    for i in range(5):
        time.sleep(1)
        print(f"Thread 1: {i}")

def print_letters():
    for letter in 'ABCDE':
        time.sleep(1)
        print(f"Thread 2: {letter}")

# Create two threads
thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_letters)

# Start the threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

print("Both threads have finished.")

```



Q2. Why threading module used? Write the use of the following functions
1. activeCount
2. currentThread
4. enumerate()


```python
print(''' The threading module in Python is used for creating, managing, 
and synchronizing threads in a program. It provides a high-level interface for working with threads, 
allowing developers to implement concurrent and parallel execution in their applications. 
''')

print(''' 1] activeCount() Function:
    Use:
        The activeCount() function is used to obtain the current number of Thread objects that are currently alive.
        This includes both daemon and non-daemon threads.
    Example: ''')

import threading

def my_function():
    print("Thread execution")

# Create two threads
thread1 = threading.Thread(target=my_function)
thread2 = threading.Thread(target=my_function)

# Start the threads
thread1.start()
thread2.start()

# Print the number of active threads
print(f"Number of active threads: {threading.active_count()}")


print('''\n currentThread() Function:
     Use:
        The currentThread() function returns the current Thread object corresponding to the caller's thread of control.
    Example: ''')
import threading

def my_function():
    current_thread = threading.current_thread()
    print(f"Thread name: {current_thread.name}")

# Create two named threads
thread1 = threading.Thread(target=my_function, name="Thread 1")
thread2 = threading.Thread(target=my_function, name="Thread 2")

# Start the threads
thread1.start()
thread2.start()


print('''\n enumerate() Function:
    Use:
        The enumerate() function returns a list of all Thread objects currently alive. It includes both daemon and non-daemon threads.
    Example: ''')

import threading

def my_function():
    print("Thread execution")

# Create two threads
thread1 = threading.Thread(target=my_function)
thread2 = threading.Thread(target=my_function)

# Start the threads
thread1.start()
thread2.start()

# Enumerate and print all active threads
for thread in threading.enumerate():
    print(f"Active Thread: {thread.name}")

```


Q3 Explain the following functions
 1. run
 2. start
 3. join
 4. isAlive()
 

```python
print(''' The functions you mentioned (run, start, join, and isAlive) 
are associated with the Thread class in the threading module in Python.
Let's discuss each of them: ''')

print(''' run() Method:
    Use:
        The run() method is called when the start() method is invoked on a Thread object. 
        It represents the code that will be executed in the new thread.
    Example: 
    ''')

import threading

class MyThread(threading.Thread):
    def run(self):
        print(f"Thread {self.name} is running")

# Create and start a thread
my_thread = MyThread()
my_thread.start()  # This invokes the run() method


print('''\n start() Method:
    Use:
        The start() method is used to initiate the execution of the thread. It invokes the run() method in a separate thread of control. It should be called only once for each thread.
    Example:
    ''')
import threading

def my_function():
    print("Thread execution")

# Create and start a thread
my_thread = threading.Thread(target=my_function)
my_thread.start()  # This invokes the run() method in a separate thread

print('''\n join([timeout]) Method:
    Use:
        The join() method is used to wait for the thread to complete its execution before moving on with the rest of the program. It blocks the calling thread until the thread whose join method is called completes.
    Example: 
    ''')

import threading

def my_function():
    print("Thread execution")

# Create and start a thread
my_thread = threading.Thread(target=my_function)
my_thread.start()

# Wait for the thread to complete
my_thread.join()
print("Main thread continues after the joined thread has finished")


print(''' \nisAlive() Method:
    Use:
        The isAlive() method is used to check if the thread is still alive (i.e., if it has been started and has not yet completed its execution).
    Example:''')
import threading
import time

def my_function():
    time.sleep(2)

# Create and start a thread
my_thread = threading.Thread(target=my_function)
my_thread.start()

# Check if the thread is still alive
while my_thread.is_alive():
    print("Thread is still running...")
    time.sleep(1)

print("Thread has completed its execution")

```


Q4. Write a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes.


```python
import threading

def print_squares():
    for i in range(1, 6):
        print(f"Thread 1: Square of {i} is {i**2}")

def print_cubes():
    for i in range(1, 6):
        print(f"Thread 2: Cube of {i} is {i**3}")

# Create two threads
thread1 = threading.Thread(target=print_squares)
thread2 = threading.Thread(target=print_cubes)

# Start the threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

print("Both threads have finished.")
```


Q5. State advantages and disadvantages of multithreading.


```python
print('''Advantages of Multithreading:

    Parallelism: Multithreading allows multiple threads to execute concurrently, taking advantage of multiple CPU cores.
    This can lead to improved performance for parallelizable tasks.

    Responsiveness: 
    Multithreading is beneficial for applications that require responsiveness. For example, in GUI applications, a separate thread can handle user input, 
    ensuring that the interface remains responsive even when the main thread is busy with computations.

    Resource Sharing:
    Threads within the same process share the same address space, making it easier to share data and resources.
    This can simplify communication between different parts of the program.

    Efficient Task Execution:
    Multithreading is useful for tasks that involve waiting for external resources, such as I/O operations. 
    While one thread is waiting, other threads can continue executing, making better use of the CPU.

    Improved Throughput:
    In certain scenarios, multithreading can lead to increased throughput by allowing a program to execute more tasks simultaneously.\n
    ''')

print('''\n Disadvantages of Multithreading:

    Complexity: 
    Multithreading introduces complexity into the program, making it harder to reason about and debug.
    Synchronization and coordination between threads can lead to subtle bugs, such as race conditions and deadlocks.

    Synchronization Overhead:
    Threads sharing data need synchronization mechanisms to avoid data corruption. 
    This synchronization overhead can impact performance and may introduce bottlenecks.

    Difficulty in Debugging:
    Debugging multithreaded programs can be challenging due to the non-deterministic nature of thread execution.
    Issues such as race conditions may not always manifest consistently.

    Potential for Deadlocks:
    Poorly managed synchronization between threads can lead to deadlocks, where threads are stuck waiting for each other to release resources.
    This can result in a program becoming unresponsive.

    Increased Memory Usage:
    Each thread has its own stack and thread-specific data, contributing to increased memory usage. 
    In certain scenarios, the overhead of creating and managing threads may outweigh the benefits.

    Global Interpreter Lock (GIL):
    In the CPython implementation (the default Python interpreter),
    the Global Interpreter Lock (GIL) restricts true parallel execution of threads.
    This limitation can impact the performance of CPU-bound tasks in multithreaded Python programs.  ''')
```


Q6. Explain deadlocks and race conditions


```python
print(''' Deadlocks:

A deadlock is a situation in concurrent programming where two or more threads are unable to proceed 
because each is waiting for the other to release a resource. In other words, each thread holds a resource 
and is also waiting for another resource acquired by another thread, creating a circular waiting condition.
As a result, the threads are effectively stuck, and the program cannot progress.

The four necessary conditions for a deadlock to occur are:

    Mutual Exclusion:
    At least one resource must be held in a non-sharable mode. Only one thread can use the resource at a time.

    Hold and Wait:
    A thread holds a resource and is waiting for another resource.

    No Preemption: 
    Resources cannot be forcibly taken away from a thread; they must be released voluntarily.

    Circular Wait: 
    There is a circular chain of two or more threads, each holding a resource needed by the next one in the chain. 
    
    ''')

print(''' Race Conditions:

A race condition occurs in concurrent programming when the behavior of a program depends on the relative timing of events,
such as the order of execution of operations on shared resources. It arises when multiple threads access shared data concurrently, 
and the final outcome depends on the interleaving of their executions.
Race conditions can result in unpredictable and undesired behavior,
such as data corruption or program crashes. They are particularly challenging to identify 
and fix because they often depend on the timing and execution order, making them difficult to reproduce consistently.

Common situations that can lead to race conditions include:

    Read-Modify-Write Operations:
    When multiple threads read, modify, and write shared data without proper synchronization, the final state can be unpredictable.

    Non-Atomic Operations:
    Operations that seem like a single operation but are implemented 
    as multiple steps can be interrupted by other threads, 
    leading to inconsistencies ''')
```
