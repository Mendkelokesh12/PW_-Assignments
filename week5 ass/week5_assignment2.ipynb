Q1. Explain why we have to use the Exception class while creating a Custom Exception.

Note: Here Exception class refers to the base class for all the exceptions.

```python
print(''' When creating a custom exception in Python, it is recommended to inherit from the Exception class or one of its subclasses. 
The Exception class is the base class for all built-in exceptions in Python, making it the natural choice for creating custom exceptions. 
Here are a few reasons why it is beneficial to use the Exception class as the base class for custom exceptions:
Consistency with the Exception Hierarchy , Compatibility with Exception Handling , Interoperability with Libraries and Frameworks .\n''')

class CustomError(Exception):
    """Custom exception inheriting from the base Exception class."""
    pass

# Using the custom exception
try:
    raise CustomError("This is a custom exception.")
except CustomError as e:
    print(f"Caught an exception: {e}")
```


Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.


```python
print(''' The ArithmeticError class is the base class for exceptions that arise during arithmetic operations.
It serves as a general category for various arithmetic-related exceptions.
Two notable exceptions that are subclasses of ArithmeticError are ZeroDivisionError and OverflowError. \n''')

try:
    result = 10 / 0  # This will raise ZeroDivisionError
except ZeroDivisionError as e:
    print(f"Error: {e}")

try:
    result = 2 ** 1000  # This may raise OverflowError for very large exponents
except OverflowError as e:
    print(f"Error: {e}")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
my_list = [1, 2, 3, 4, 5]

try:
    # Some code that may raise KeyError or IndexError
    value = my_dict['x']  # This may raise KeyError
    value = my_list[10]   # This may raise IndexError

except KeyError as e:
    print(f"Error: {e}")
    
except IndexError as e:
    print(f"Error: {e}")

except LookupError as e:
    print(f"LookupError: {e}")
```


Q5. Explain ImportError. What is ModuleNotFoundError ?


```python
print(''' ImportError is a base class for exceptions that occur when importing a module.
It is raised when there is an issue with the import statement or 
when the Python interpreter encounters difficulties while trying to import a module. ImportError itself is a general exception,
and there are more specific exceptions that inherit from it, such as ModuleNotFoundError
''')

try:
    import non_existent_module  # This will raise ImportError
except ImportError as e:
    print(f"ImportError: {e}")
```

```python
print(''' ModuleNotFoundError: A more specific subclass of ImportError introduced in Python 3.6. 
It is raised when the specified module cannot be found.
Example:
''')

try:
    from non_existent_package import some_module  # This will raise ModuleNotFoundError
except ModuleNotFoundError as e:
    print(f"ModuleNotFoundError: {e}")
```
